```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn import metrics
import pydotplus
from sklearn.model_selection import train_test_split
```

## Step 1:
Use the "titanic" dataset in homework 3

```python
#Reading in Data
df = pd.read_csv('Titanic.csv')
df.head()
```

```python
#Dropping unnecessary columns
df.drop(df.columns[[0, 3, 7, 8, 9, 10, 11, 12, 13, 14]], axis=1, inplace=True)
```

## Step 2:
Fit a neural network using independent variables ‘pclass + sex + age + sibsp’ and dependent variable ‘survived’. Fill in n/a attributes with the average of the same attributes from other training examples.

```python
#Fill Data with averages using fillna
#Used numeric_only=True to only apply code to columns/rows with numerical values
df.fillna(value=df.mean(numeric_only=True))
```

```python
features = ['pclass', 'sex', 'age', 'sibsp']
X = df[features].apply(LabelEncoder().fit_transform) 
y = df['survived']

print(X)
print(y) 
```

```python
#Loading additional libraries & Modules
from sklearn.neural_network import MLPClassifier
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error
from math import sqrt
from sklearn.metrics import r2_score
```

```python
print(df.shape)
```

```python
df.describe().transpose()
```

The mean value for survived is 0.38, meaning that ~38% survived. Therefore, the baseline accuracy is 62%, which our neural network model should exceed. 

```python
#Training the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
print(X_train.shape)
print(X_test.shape)
```

Use 2 hidden layers and set the activation functions for both the hidden and output layer to be the sigmoid ('logistic') function. Set “solver” parameter as either SGD (stochastic gradient descend) or Adam (similar to SGD but optimized performance with mini batches). You can adjust parameter “alpha” for regularization (to control overfitting) and other parameters such as “learning rate” and “momentum” as needed

```python
#Building, Predicting, and Evaluating the Neural Network Model
from sklearn.neural_network import MLPClassifier

mlp = MLPClassifier(hidden_layer_sizes=(16, 8), activation='logistic', solver='adam', alpha=1e-5, momentum=0.1, max_iter=300, verbose=True)
mlp.fit(X_train, y_train)

predict_train = mlp.predict(X_train)
predict_test = mlp.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(y_train,predict_train))
print(classification_report(y_train,predict_train))
```

```python
print(confusion_matrix(y_test,predict_test))
print(classification_report(y_test,predict_test))
```

Check the performance of the model with out-of- sample accuracy, defined as;

out-of-sample percent survivors correctly predicted (on test set)

out-of-sample percent fatalities correctly predicted (on test set)

```python
outofsamplesurvivors = 1 - sum(predict_test[y_test ==1]^y_test[y_test == 1])/len(y_test[y_test ==1])
outofsamplefatalities = 1 - sum(predict_test[y_test ==0]^y_test[y_test == 0])/len(y_test[y_test ==0])

print(f"Correctly Predicted Out-of-Sample percent Survivors: {outofsamplesurvivors*100:.2f}%")
print(f"Correctly Predicted Out-of-Sample percent Fatalities: {outofsamplefatalities*100:.2f}%")
```
